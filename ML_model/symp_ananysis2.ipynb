```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('../data/diseasedataf.csv')
df.head()
```

```python
df.fillna(method='ffill', inplace=True)
df
```

```python
df.Disease=df.Disease.str.replace('^','_').str.split('_').str[1]
df.Symptom=df.Symptom.str.replace('^','_').str.split('_').str[1]
df
```

```python
p=df.groupby(['Disease'])
```

```python
symp=df['Symptom'].unique()
dis=df['Disease'].unique()
c=df['Count'].unique()
symp,dis,c
```

```python
len(df['Symptom'].unique()) , len(df['Disease'].unique()) , len(df['Count'].unique())
```

```python
df1=pd.DataFrame(columns=symp)
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
arr_le=le.fit_transform(df['Symptom'])
arr_le
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer 
ohe=OneHotEncoder(sparse=False)
arr_ohe=ohe.fit_transform(arr_le.reshape(len(arr_le),1))
arr_ohe
```

```python
for i in range(len(arr_ohe)):
    df1.loc[i]=arr_ohe[i]
df1
```

```python
df.drop(columns='Symptom',inplace=True)
df.head(), df.shape , df1.shape
```

```python
df=pd.concat([df,df1],axis=1)
df, df.shape
```

```python
df.drop_duplicates(keep='first',inplace=True)
df.shape
```

```python
df=df.groupby(['Disease','Count']).sum()
df.head()
```

```python
df=df.reset_index()
df.head()
```

```python
df.loc[3]
```

```python
df.corr()
```

```python
df.drop(columns=['Count'],inplace=True)
```

```python
X=df.drop(columns=['Disease'])
Y=df['Disease']
X.head(),Y.head()
```

```python
# from sklearn.model_selection import train_test_split
# X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
# X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import tree
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
def train_model(model):
    at=model.fit(X, Y)
    score=at.score(X, Y)
    return score
```

```python
model=[DecisionTreeClassifier(),SVC(),LogisticRegression(),GaussianNB(),KNeighborsClassifier(n_neighbors=21),RandomForestClassifier(n_estimators=101)]
score=[]
for i in model:
    x=train_model(i)
    score.append(x)
result=dict(zip(model,score))
result
```
